```python
from TissueMask import TissueMask
from TissueSlide import TissueSlide
import matplotlib.pyplot as plt
def visualize(*images):
    n = len(images)
    fig, axes = plt.subplots(1, n, figsize=(5 * n, 5))  

    if n == 1:
        axes = [axes]  

    for i, (img, title) in enumerate(images):
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(title) 

    plt.tight_layout()
    plt.show()
```

Green Pen
==

```python
#green pen example
green_path = "./EXAMPLE/TCGA-A1-A0SH-01Z-00-DX1.90E71B08-E1D9-4FC2-85AC-062E56DDF17C.svs"
green_pen_slide = TissueSlide(green_path)
visualize((green_pen_slide.thumbnail, "Thumbnail"))
```

```python
whole_tissue = TissueMask(green_pen_slide, masks =["whole_slide"])
mask_green = TissueMask(green_pen_slide, masks =["green_pen"])
mask_red = TissueMask(green_pen_slide, masks =["red_pen"])
mask_blue =TissueMask(green_pen_slide, masks =["blue_pen"])
mask_all = TissueMask(green_pen_slide)
```

```python
visualize((whole_tissue.mask, "Only Tissue"), (mask_green.mask, "excluding only green pen"),
          (mask_red.mask, "excluding only red pen"), (mask_blue.mask, "excluding only blue pen"), 
          (mask_all.mask, "all masked"))
```

```python
visualize((whole_tissue.applied, "Only Tissue"), (mask_green.applied, "excluding only green pen"),
          (mask_red.applied, "excluding only red pen"), (mask_blue.applied, "excluding only blue pen"),
         (mask_all.applied, "all masked"))
```
